```python
"""
# WHAT IS Machine Learning??
*The Subfield of computer science that gives computers the ability to learn (pattern in the data)
without explicitly programs.

- Machine learns patterns from the data and replicates the same in future.

*A Computer program is said to learn from experience E with respect to some class of
Task T and Performance measure P if its performance at Task T as measured by P
improves with experience E

Task(T) : classifying an email into Spam/Ham
Experience(E) : watching your label of email
Performance Measure(P) : the no. of emails correctly classified

# Why ML is famous these Days?
-> Advance Processor Nowdays which help us to compute the models
-> Data is new oil 
( A lot of data has been created or generated 
-> so we process this data using advance processor
-> valuable insights
-> which improves business performance) 

#ML Styles:
-> Supervised Learning
-> Unsupervised Learning
-> Semi - Supervised Learning
-> Reinforcement Learning
-------------------------------------------------------------------------------------
#AI(Artificial Intelligence) vs
 ML(Machine Learning) vs 
 DL(Deep Learning) vs 
 DS(Data Science)

AI(Artificial Intelligence): describes how computer and other technology mimics human intelligence
(smart machines that can perform a task without human intervention)
EX: Robots , Self-driving Cars

ML(Machine Learning): that focus on creating algorithms and statistical models to let 
computers learn and make predictions without explicitly programmed
EX: classifying an email a Spam or Ham ,
    predicition of a person if he is diabetic or non diabetic,
    price of house
Real world applications - Amazon recommendation system
                          Netflix recommendation system
                          
DL(Deep Learning): Specialised Ml algorithms that mimics human brain 
(Multilayered Neural Network)
EX: Chatgpt
    Recommendation
    Object detection
    Image Recognition

DS(Data Science): Inter-Disciplinary field that comprises of CS,Stats,ML and Domain Knowledge

-------------------------------------------------------------------------------------

#Types of ML :- 4 types
-> Supervised Learning
-> Unsupervised Learning
-> Semi-Supervised Learning
-> Reinforcement Learning

1) Supervised Learning: The Machine Learning where the model is trained on Labelled Data.
                        The goal is to learn a mapping function to predict outputs for new, unseen data.

2) Unsupervised Learning : The Machine Learning where the model is trained on Labelled Data.
                           It identifies hidden patterns or structures without predefined outputs.

3) Semi-Supervised Learning : A hybrid approach which uses small amount of labelled data
                              and a large amount of unlabelled data.
                              It improves learning accuracy while reducing labelling costs.

4) Reinforcement Learning : A learning Technique where an agents learns by interacting with an
                            enviornment and receives rewards based on its actions
                            Aiming to maximize total reward overtime.

-------------------------------------------------------------------------------------

#Train , Validation and Test Data:

Train Data : training the model (EX: teaching a student)
Validation : Hyperparameter tuning of model( EX: Sample papers solving)
Test Data : Test the models accuracy (EX: like final exams)

*In a 100% data we divide the data into 3 parts
EX: 60% in training , 20% in validation , 20% to test performance of model

-------------------------------------------------------------------------------------

# Model Training: Given the data estimate the prediction function by minimising the error.
-> Model performance good on test data (Accuracy should be high)
    if accuracy is low then it is a problem

# Model overfitting :
Train data -> Model is trained -> Accuracy is high (95%)
Test data(Unseen Data) -> Model is tested -> Accuracy is very low (65%)

Overfitting? - Model performs well on trained data but worst on test data

# Model underfitting :
Train data -> Model is trained -> Accuracy is low (50%)
Test data(Unseen Data) -> Model is tested -> Accuracy is very low (30%)

Underfitting? - Model performs low on trained data but worst on test data

#Generalised Model: 
Train data -> Model is trained -> Accuracy is high (85%)
Test data(Unseen Data) -> Model is tested -> Accuracy is high (84%)

Generalised Model? - Model performs well on trained data but also well on test data

-------------------------------------------------------------------------------------

# Bias / Variance:
* Training error is also known as bias. 
               -> High training error means high bias
* Testing error is also known as variance.
               -> High testing error means high variance 

#Bias / Variance trade off:


-------------------------------------------------------------------------------------

#Handling Missing Values :
-> Reasons for missing data:
                            1)Human error
                            2)Data Corrupt
                            
*Missing data occurs in a dataset when some of the information is not saved / stored

(1) Missing completely at random (MCAR):
-> Data missing is independent of observed and unobserved data.
*Survey : if a person is suffering from a disease or not .

(2) Missing at Random :
-> The missing data depends on observed data but not on the missing data itself.
*Survey : to know the income of people

(3) Missing data not at random :
-> Missing values depends on the value of missing data itself
*Survey : income and job satisfaction of employees in a company


-------------------------------------------------------------------------------

* Missing value treatment (should be asked from business team) -
Why? - For analysis and ml prediction it is important to deal with missing values

    1) if the data us huge and missing value is less than 1% of data , you can drop the missing values.
    2) impute(Replace) the missing values
    
    If continuous we use mean and median | If categorical we use mode
    
*If the outlier treatment is not done than impute the missing value with median.
   
    3) if any column has missing value greater than 40% than drop the column
    4) impute or replace the missing value with a random number or with a value which is not possible
    
    -------------------------------------------------------------------------------------
    """
```
